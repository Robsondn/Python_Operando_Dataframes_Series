## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
#1) Meu codigo
percent_mat = df_mat / 24 * 100
percent_por = df_por / 18 * 100
percent_geo = df_geo / 8 * 100
percent_ing = df_ing / 8 * 100
percent_his = df_his / 8 * 100
percent_fis = df_fis / 12 * 100
percent_qui = df_qui / 12 * 100

print("Percentual de acertos em Matemática:")
print(percent_mat)
print("\nPercentual de acertos em Português:")
print(percent_por)
print("\nPercentual de acertos em Geografia:")
print(percent_geo)
print("\nPercentual de acertos em Inglês:")
print(percent_ing)
print("\nPercentual de acertos em História:")
print(percent_his)
print("\nPercentual de acertos em Física:")
print(percent_fis)
print("\nPercentual de acertos em Química:")
print(percent_qui)
````

````python
# 2) Meu código
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
print("Total de acertos por aluno:")
print(total_acertos)
````

````python
# 3) Meu código
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
total_questoes = 24 + 18 + 8 + 8 + 8 + 12 + 12
percentual_geral = (total_acertos / total_questoes) * 100

print("Percentual geral de acertos por aluno:")
print(percentual_geral)
````

````python
# 4) Meu código
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
alunos_aprovados = total_acertos[total_acertos['Qt_acertos'] > 45]
num_alunos_aprovados = len(alunos_aprovados)

print("Número de alunos com nota maior que 45:", num_alunos_aprovados)
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
 #Meu código: Critério para desqualificação: zerar em matemática, física ou química
desqualificados = (df_mat['Qt_acertos'] == 0) | (df_fis['Qt_acertos'] == 0) | (df_qui['Qt_acertos'] == 0)

# 1) Criar um DataFrame com os alunos desqualificados
df_alunos_desqualificados = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)[desqualificados]

print("DataFrame com os alunos desqualificados:")
print(df_alunos_desqualificados)

# 2) Contar o número de alunos desqualificados
num_desqualificados = len(df_alunos_desqualificados)
print(f"\nNúmero de alunos desqualificados: {num_desqualificados}")

# 3) Calcular a média de acertos dos alunos desqualificados em história e geografia
media_his_geo_desqualificados = df_alunos_desqualificados[['Qt_acertos']].iloc[:, [4, 2]].mean()
print("\nMédia de acertos dos alunos desqualificados em história e geografia:")
print(media_his_geo_desqualificados)

# Calcular o total de acertos para todos os alunos
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
# 4) Criar um DataFrame com os alunos que passaram para a segunda fase
alunos_aprovados = total_acertos[(total_acertos['Qt_acertos'] > 45) & ~desqualificados]

print("\nDataFrame com os alunos que passaram para a segunda fase:")
print(alunos_aprovados)
````
